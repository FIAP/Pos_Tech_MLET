Operadores Aritméticos:

Soma (+), subtração (-), multiplicação (*), divisão (/), divisão inteira (//), módulo (%) e exponenciação (**).

```python
a = 10
b = 3
print("Soma:", a + b)         # 13
print("Divisão Inteira:", a // b)  # 3
print("Exponenciação:", a ** b)    # 1000
```

Operadores Relacionais:

Comparam valores e retornam resultados booleanos (True ou False), utilizando ==, !=, >, <, >= e <=.

```python
x = 5
y = 8
print(x < y)  # True
```

Operadores Lógicos:

Permitem combinar condições com and, or e not.

```python
idade = 25
tem_carteira = True
print(idade > 18 and tem_carteira)  # True
```

Operadores de Atribuição:

Atualizam variáveis de forma compacta com =, +=, -=, etc.

```python
contador = 10
contador += 5
print(contador)  # 15
```

Estruturas Condicionais

if: Verifica uma condição e, se verdadeira, executa o bloco associado.

elif: Permite testar condições alternativas.

else: Executa um bloco padrão quando nenhuma condição anterior é satisfeita.


```python
nota = 8
if nota >= 7:
    print("Aprovado")
elif nota >= 5:
    print("Recuperação")
else:
    print("Reprovado")
```

Estruturas Aninhadas

Você pode combinar condições e criar estruturas aninhadas para resolver situações mais complexas.

```python
idade = 20
tem_carteira = True
if idade >= 18:
    if tem_carteira:
        print("Pode dirigir.")
    else:
        print("Precisa de carteira para dirigir.")
else:
    print("Não pode dirigir.")
```

Operadores Lógicos em Condicionais

Combinar condições com and, or para testes mais robustos.

```python
idade = 22
experiencia = 3
if idade >= 21 and experiencia >= 2:
    print("Elegível para a vaga.")
else:
    print("Não elegível para a vaga.")
```
