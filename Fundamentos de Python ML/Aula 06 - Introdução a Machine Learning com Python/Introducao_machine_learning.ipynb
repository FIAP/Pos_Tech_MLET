NumPy – Operações Numéricas com Arrays
Objetivo: Aprender a criar e manipular arrays, que são a base para operações numéricas eficientes.

1 - Importação e Criação de Arrays:

```python
import numpy as np

# Criando um array a partir de uma lista
lista = [1, 2, 3, 4, 5]
array1 = np.array(lista)
print("Array1:", array1)

# Criando um array 2D (matriz)
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print("Matriz:\n", matriz)
```

2 - Operaçõs Básicas com Arrays:


```python
# Soma, multiplicação e operações element-wise
array2 = np.array([10, 20, 30, 40, 50])
soma = array1 + array2
multiplicacao = array1 * array2
print("Soma:", soma)
print("Multiplicação:", multiplicacao)
```

3 - Funções NumPy Comuns: 

```python
media = np.mean(array2)
desvio = np.std(array2)
print("Média:", media)
print("Desvio padrão:", desvio)
```

 Pandas – Manipulação e Análise de Dados
 
Objetivo: Utilizar o Pandas para carregar, explorar e manipular conjuntos de dados de forma eficiente.

1 - Importação e Criação de DataFrame:

```python
%pip install pandas
```

```python
import pandas as pd

# Criando um DataFrame a partir de um dicionário
dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos'],
    'Idade': [25, 30, 35],
    'Salário': [5000, 7000, 8000]
}
df = pd.DataFrame(dados)
print("DataFrame:\n", df)
```

Manipulação e Exploração de Dados:

```python
# Selecionando uma coluna
print("Coluna 'Nome':\n", df['Nome'])

# Filtrando linhas com base em uma condição
df_filtrado = df[df['Salário'] > 6000]
print("Funcionários com salário > 6000:\n", df_filtrado)

# Estatísticas descritivas
print("Estatísticas:\n", df.describe())
```

Leitura de Dados a partir de Arquivos CSV:

```python
# Suponha que você tenha um arquivo 'dados.csv'
df_csv = pd.read_csv('dados.csv')
print(df_csv.head())
```

Matplotlib – Visualização de Dados
Objetivo: Aprender a criar gráficos para visualizar os dados de maneira clara e informativa.

Importação e Criação de Gráficos Simples:

```python
%pip install matplotlib
```

```python
import matplotlib.pyplot as plt

# Gráfico de linha
plt.plot(df['Idade'], df['Salário'], marker='o')
plt.title("Salário x Idade")
plt.xlabel("Idade")
plt.ylabel("Salário")
plt.show()
```

Personalização de Gráficos:

```python
# Gráfico de barras
plt.bar(df['Nome'], df['Salário'], color='skyblue')
plt.title("Salário por Nome")
plt.xlabel("Nome")
plt.ylabel("Salário")
plt.show()
```

Seaborn – Visualizações Avançadas
Objetivo: Utilizar o Seaborn para criar gráficos estatísticos com estética aprimorada e insights aprofundados.

Importação e Configuração Básica:

```python
%pip install seaborn
```

```python
import seaborn as sns

# Configuração padrão para gráficos mais bonitos
sns.set(style="whitegrid")
```

Criação de Gráficos com Seaborn:

```python
# Gráfico de dispersão (scatter plot)
sns.scatterplot(x='Idade', y='Salário', data=df, hue='Nome', s=100)
plt.title("Scatter Plot: Salário vs. Idade")
plt.show()
```

```python

# Boxplot para visualizar a distribuição do salário
sns.boxplot(y='Salário', data=df)
plt.title("Distribuição de Salário")
plt.show()
```

Integração com Pandas: O Seaborn trabalha diretamente com DataFrames do Pandas, facilitando a criação de gráficos que refletem as estatísticas dos dados.

```python
# Gráfico de distribuição (histograma com KDE)
sns.histplot(df['Salário'], kde=True)
plt.title("Histograma do Salário")
plt.show()
```
