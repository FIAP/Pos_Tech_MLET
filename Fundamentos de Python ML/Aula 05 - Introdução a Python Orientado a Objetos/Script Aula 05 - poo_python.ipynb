1. Conceitos Básicos: Classes, Objetos, Atributos, Métodos e Construtores

```python
class Carro:
    def __init__(self, modelo, cor, ano):
        # O método __init__ é o construtor que inicializa os atributos do objeto.
        self.modelo = modelo
        self.cor = cor
        self.ano = ano

    def acelerar(self, velocidade):
        # Método que simula a aceleração do carro.
        return f"O {self.modelo} está acelerando a {velocidade} km/h."

# Criando um objeto da classe Carro
meu_carro = Carro("Polo", "Preto", 2025)
print(meu_carro.acelerar(120))
```

Princípios Avançados de POO

Herança:

```python
class Veiculo:
    def __init__(self, marca, ano):
        self.marca = marca
        self.ano = ano

    def descricao(self):
        return f"Veículo da marca {self.marca}, ano {self.ano}"

class Carro(Veiculo):
    def __init__(self, marca, ano, modelo):
        # Utiliza a herança para inicializar marca e ano
        super().__init__(marca, ano)
        self.modelo = modelo

    def descricao(self):
        # Polimorfismo: reescreve o método da classe base para incluir o modelo
        return f"Carro: {self.marca} {self.modelo}, ano {self.ano}"

# Exemplo de uso
meu_carro = Carro("Polo", "Preto", 2025)
print(meu_carro.descricao())
```

Polimorfismo

```python
class Animal:
    def emitir_som(self):
        pass

class Gato(Animal):
    def emitir_som(self):
        return "Miau"

class Cachorro(Animal):
    def emitir_som(self):
        return "Au au"

animais = [Gato(), Cachorro()]

for animal in animais:
    print(animal.emitir_som())  # Miau, Au au
```

Encapsulamento

```python
class ContaBancaria:
    def __init__(self, saldo_inicial):
        self.__saldo = saldo_inicial  # Atributo privado

    def depositar(self, valor):
        if valor > 0:
            self.__saldo += valor

    def obter_saldo(self):
        return self.__saldo

# Exemplo de uso
conta = ContaBancaria(1000)
conta.depositar(500)
print("Saldo:", conta.obter_saldo())
```

Aqui, o saldo é privado e só pode ser acessado por meio de get_saldo().
